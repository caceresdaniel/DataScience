## Importing all necessary sklearn methods and so on 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import scale
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
```

## Placing the data set into a pandas dataframe

```python
df = pd.read_csv('https://raw.githubusercontent.com/caceresdaniel/DataScience/master/Heart_short.csv')
df
```

## Replacing the Yes with 1 and No with 0 in the AHD feature to be able to properly work with the data set

```python
df = df.replace({'AHD': {'Yes': 1, 'No': 0}})
df
```

## Placing features and label in variables and then scaling the features

```python
feats = ['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak']
x = df[feats]

y = df['AHD']

newx = scale(x)
```

## Splitting the data set into testing and training sets with test size = 0.25 and random state = 3

```python
x_train, x_test, y_train, y_test = train_test_split(newx, y, test_size=0.25,random_state=3)
```

## Using LogisticRegression to predict heart disesase then computing the accuracy

```python
logreg = LogisticRegression()

logreg.fit(x_train, y_train)

predict = logreg.predict(x_test)

acc = accuracy_score(y_test, predict)

print(acc)
```

## Using LogisticRegression to predict the probability 

```python
prob = logreg.predict_proba(x_test)

print(prob)

print(y_test)
```

## Calculating the AUC

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test,prob[:,1], pos_label=1)

AUC = metrics.auc(fpr, tpr)

print(AUC)
```

## Ploting the ROC Curve of the data set using the AUC

```python
%matplotlib inline

plt.figure()

plt.plot(fpr, tpr, color='red', lw=2, 
         label='ROC Curve (area = %0.2f)' % AUC)

plt.plot([0, 1], [0, 1], color='blue', lw=1, linestyle='--')

plt.xlim([-0.005, 1.005])
plt.ylim([0.0, 1.01])

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")

plt.show()
```
