## Importing all necessary utilities 

```python
import numpy as np

from keras.wrappers.scikit_learn import KerasClassifier
from keras.layers import Dense, Dropout, Activation, Flatten
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.utils import np_utils
from keras.datasets import mnist
```

## Loaded the data then reshaped it, then finaly scaled it and divided by 255

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

## One hot encoded the label in to a vector of 10 elements 

```python
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

## Defined a function which defines creates and compiles the neural network model 

```python
def model_creator():
    model = Sequential()
    input_size = 784
    hidden_neurons = 100
    out_size = 10
    model.add(Dense(hidden_neurons, input_dim = input_size))  # Nuerons
    model.add(Activation('sigmoid')) # Activation
    model.add(Dense(out_size, input_dim = hidden_neurons))  # Nuerons
    model.add(Activation('softmax')) # Activation
    model.compile(loss='categorical_crossentropy',
              metrics=['accuracy'],
              optimizer='adam')
    return model
```

## Fixed the random state for reproducibility

```python
seed = 2
np.random.seed(seed)
```

## Used the keras classifier to wrap the model as an object

```python
model = KerasClassifier(build_fn = model_creator, verbose=2)
```

## Running Grid Search to find the best epoch and bachsize using 10 folds 

```python
epochs = np.array([10 , 15 , 20])
batches = np.array([30 , 50 , 100])
param_grid = dict(epochs=epochs, batch_size=batches)
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_log_loss', cv=10)
grid_result = grid.fit(X_train, y_train)
```

## Printing out the best batch size and epochs and also printing out the best score for each epoch and batch size

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
for params, mean_score, scores in grid_result.grid_scores_:
    print("%f (%f) with: %r" % (scores.mean(), scores.std(), params))
```

## Predicting on the best epoch/batch size

```python
y_predict = grid.best_estimator_.model.predict(X_test)
print (y_predict.shape)
```

## Printing out the accuracy from the best epoch/batch size

```python
score = grid.best_estimator_.model.evaluate(X_test, y_test)
print('The accuracy is: ', score[1])
```
