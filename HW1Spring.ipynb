```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.utils import resample
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/caceresdaniel/DataScience/master/Cancer.csv')
df
```

```python
feats = ["Clump_Thickness", "Uniformity_of_Cell_Size", "Uniformity_of_Cell_Shape", "Marginal_Adhesion", "Single_Epithelial_Cell_Size", "Bare_Nuclei", "Bland_Chromatin", "Normal_Nucleoli", "Mitoses"]
x = df[feats]

y = df['Malignant_Cancer']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=2)
```

```python
dt = DecisionTreeClassifier(random_state=2)

dt.fit(x_train, y_train)

dtPredict = dt.predict(x_test)

dtacc = accuracy_score(y_test, dtPredict)

print(dtacc)
```

```python
dtPredProb = dt.predict_proba(x_test)
fpr, tpr, thresholds = metrics.roc_curve(y_test, dtPredProb[:,1],pos_label=1)
AUC = metrics.auc(fpr,tpr)
print(AUC)
```

```python
preds = []

for i in range(0, 29):
    bootstarp_size = int(0.8*(x_train['Clump_Thickness'].count()))
    x_samp, y_samp = resample(x_train,y_train, n_samples=bootstarp_size, random_state=i, replace=True)
    Base_DecisionTree = DecisionTreeClassifier(random_state=3)
    Base_DecisionTree.fit(x_samp,y_samp)
    pred = Base_DecisionTree.predict(x_test)
    preds.append(pred)

#Function that finds the most common element in the list
def most_common(lst):
    return max(set(lst), key=lst.count)    

vote = []
for i in range (0,45):
    temp = []
    for pred in preds:
        temp.append(pred[i])
    vote.append(most_common(temp))

acc = accuracy_score(y_test,vote)

print(acc)
```

```python
my_AdaBoost = AdaBoostClassifier(n_estimators = 29,random_state=2)
my_AdaBoost.fit(x_train, y_train)
adap = my_AdaBoost.predict(x_test)
adaacc = accuracy_score(y_test, adap)

print(adaacc)
```

```python
adaProb = my_AdaBoost.predict_proba(x_test)
adafpr, adatpr, thresholds = metrics.roc_curve(y_test, adaProb[:,1],pos_label=1)
adaAUC = metrics.auc(adafpr,adatpr)
print(adaAUC)
```

```python
my_XGBoost = XGBClassifier(n_estimators = 29,random_state=2)
my_XGBoost.fit(x_train, y_train)
xgbp = my_XGBoost.predict(x_test)
xgpacc = accuracy_score(y_test, xgbp)

print(xgpacc)
```

```python
xgbprob = my_XGBoost.predict_proba(x_test)
xgbfpr, xgbtpr, thresholds = metrics.roc_curve(y_test, xgbprob[:,1], pos_label=1)
xgbAUC = metrics.auc(xgbfpr, xgbtpr)
print(xgbAUC)
```

```python
my_RandomForest = RandomForestClassifier(n_estimators = 29, bootstrap = True, random_state=2)
my_RandomForest.fit(x_train, y_train)
rfp = my_RandomForest.predict(x_test)
rfacc = accuracy_score(y_test, rfp)

print(rfacc)
```

```python
rfprob = my_RandomForest.predict_proba(x_test)
rffpr, rftpr, thresholds = metrics.roc_curve(y_test, rfprob[:,1], pos_label=1)
rfAUC = metrics.auc(rffpr, rftpr)
print(rfAUC)
```

```python
%matplotlib inline

plt.figure()

plt.plot(fpr, tpr, color='red', lw=2,
        label='Decision Tree (area = %0.2f)' % AUC)
plt.plot(fpr, tpr, color='purple', lw=2,
        label='Bagging (area = %0.2f)' % AUC)
plt.plot(adafpr, adatpr, color='yellow', lw=2,
        label='AdaBoost (area = %0.2f)' % adaAUC)
plt.plot(xgbfpr, xgbtpr, color='black', lw=2,
        label='XGBoost (area = %0.2f)' % xgbAUC)
plt.plot(rffpr, rftpr, color='pink', lw=2,
        label='Random Forest (area = %0.2f)' % rfAUC)

plt.plot([0, 1], [0, 1], color='blue', lw=1, linestyle='--')

plt.xlim([-0.005, 1.005])
plt.ylim([0.0, 1.01])

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Cancer Predicton')
plt.legend(loc="lower right")

plt.show()
```
